```python
mid_sal_min = df_new['Mid-Career Median Salary'].idxmin()
df_new.loc[mid_sal_min]
```

```python
spre_col = df_new['Mid-Career 90th Percentile Salary'] - df_new['Mid-Career 10th Percentile Salary']
df_new.insert(1, 'gap', spre_col)
df_new.head()
```

```python
low_risk = df_new.sort_values('gap')
low_risk[['Undergraduate Major', 'gap']].head()
```

```python
# 잠재력이 높은 전공
high_poten = df_new.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_poten[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# 소득 격차가 큰 전공
high_gap = df_new.sort_values('gap', ascending=False)
high_gap[['Undergraduate Major', 'gap']].head()
```

```python
# 전공을 3가지 범주로 묶기
df_new.groupby('Group').count()
```

```python
# 그룹별 평균 봉급
pd.options.display.float_format = '{:,.2f}'.format
df_new.groupby('Group').mean()
```
